```python
from transformers import AutoFeatureExtractor, AutoModelForImageClassification
from PIL import Image

extractor = AutoFeatureExtractor.from_pretrained("therealcyberlord/stanford-car-vit-patch16")
model = AutoModelForImageClassification.from_pretrained("therealcyberlord/stanford-car-vit-patch16")
```

```python
import requests
url = 'https://avatars.mds.yandex.net/get-autoru-vos/2080053/067a18b2f9b507426e6f177a6beab25f/1200x900n'
image = Image.open(requests.get(url, stream=True).raw)
from PIL import Image
import matplotlib.pyplot as plt


plt.imshow(image)
plt.axis('off') 
plt.show()
```

Rolls-Royce Ghost, 2012

https://auto.ru/cars/used/sale/rolls_royce/ghost/1125052555-e0870d09/?geo_id=20692

```python
import requests
inputs = extractor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits

predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```

```python
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor2 = AutoImageProcessor.from_pretrained("google/vit-base-patch16-224")
model2 = AutoModelForImageClassification.from_pretrained("google/vit-base-patch16-224")
```

```python
url = 'https://the-cafe.ru/wp-content/uploads/2020/11/0777b9cf-c854-42e5-87b8-f6988f31cfac.jpg'
image = Image.open(requests.get(url, stream=True).raw)

plt.imshow(image)
plt.axis('off') 
plt.show()
```

```python
inputs = processor2(images=image, return_tensors="pt")
outputs = model2(**inputs)
logits = outputs.logits

predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model2.config.id2label[predicted_class_idx])
```
